```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
# hitting stats csv
hitting_stats = pd.read_csv("yankees_hitters_21.csv")
hitting_stats.head()
```

```python
# list all column headers
list(hitting_stats)
```

```python
# show dataframe with all columns displayed
pd.options.display.max_columns = None
display(hitting_stats)
```

```python
# group players by position
outfielders = hitting_stats[(hitting_stats["Pos"] == "OF")]
outfielders
```

```python
# group by each infield position
first_baseman = hitting_stats[(hitting_stats["Pos"] == "1B")]
second_baseman = hitting_stats[(hitting_stats["Pos"] == "2B")]
third_baseman = hitting_stats[(hitting_stats["Pos"] == "3B")]
shortstops = hitting_stats[(hitting_stats["Pos"] == "SS")]
```

```python
# Concatenate frames to create dataframe for infielders
frames = [first_baseman, second_baseman, third_baseman, shortstops]
infielders = pd.concat(frames)
infielders
```
